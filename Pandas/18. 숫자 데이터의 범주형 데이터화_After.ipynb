#### Age 변수 변환하기
 - 변환 로직을 함수로 만든 후, apply 함수로 적용

```python
import math  # 소수점 버리려고 math 사용함
```

```python
def age_categorize(age):   #예를 들어 23 -> 20, 38 -> 30 으로 바뀜
    if math.isnan(age):   #age 가 NaN이면 -1 반환
        return -1
    return math.floor(age / 10) * 10
```

```python
train_data
```

```python
train_data['Age'].apply(age_categorize) # 바뀐 데이터 확인 가능
```
